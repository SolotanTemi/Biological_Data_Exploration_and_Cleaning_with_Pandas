```python
# This to display all rows and columns without truncation
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Importing the data file
bacteria_path = 'https://raw.githubusercontent.com/HackBio-Internship/2025_project_collection/main/bacteria.csv'

# Reading the CSV file into a DataFrame
bacteria_data = pd.read_csv(bacteria_path)
bacteria_data.head(5)
```

```python
# PART 1: CONFIMRING THE DATASET
# Printing the shape of the dataset (number of rows and columns)
print('Shape:') 
bacteria_data.shape
```

```python
# Checking the list of all column names
print('Columns:')
bacteria_data.columns.tolist()
```

```python
# PART 2: DATA SELECTION
# Selecting the 15th row of the dataset using integer location
print('Row 15:')
bacteria_data.iloc[14]
```

```python
# Selecting an entire column by its name 'species'
print('Column species:')
bacteria_data['species']
```

```python
# Retrieve the specific value located at the 15th row of the 'Isolation Origin' column
print('Value at row 15, Isolation origin:')
bacteria_data.loc[14, 'Isolation origin']
```

```python
# PART 3: DATA FILTERING
# Creating a new dataframe containing only the rows where carb_fit > 0.5
filtered_data = bacteria_data[bacteria_data['carb_fit'] > 0.5]

print('Dataframe with rows where carb_fit > 0.5:')
display(filtered_data)
```

```python
# PART 4: HANDLING MISSING DATA
# Creating a copy of the bacteria_data to use for cleaning exercises
bacteria_data_missing = bacteria_data.copy()

# Counting the number of missing (NaN) values per row
print('Number of missing values per row:')
bacteria_data_missing.isna().sum(axis=1)
```

```python
# Creating a new version of the dataset where all missing values are filled with the number 1
df_filled = bacteria_data_missing.fillna(1)
print('Filled Dataset:')
display(df_filled)
```

```python
# Creating a new version of the dataset where any column containing at least one missing value is dropped entirely
df_dropped_cols = bacteria_data_missing.dropna(axis=1)
display(df_dropped_cols)
```

```python
# Checking if there are any missing values in the main dataset (returns True or False)
print(bacteria_data.isna().any().any())

print('Missing values per column in bacteria_data:')
bacteria_data_missing.isna().sum()
```

```python
# To report the total number of missing values across the entire dataset
print('Total missing values:', bacteria_data_missing.isna().sum().sum())
```

```python
# PART 5: COLUMN RENAMING
# Renaming columns 'C1' and 'C2' to 'UMAP1' and 'UMAP2'
# A new dataframe (bacteria_data_renamed) is created to preserve the original dataset (bacteria_data).
bacteria_data_renamed = bacteria_data.rename(columns={"C1": "UMAP1", "C2": "UMAP2"})

# Displaying the first 5 rows to confirm the column names were successfully changed
print('Renamed Dataset (first 5 rows):')
display(bacteria_data_renamed.head())
```

```python
# PART 6: SORTING
# Sorting bacteria_data by the BSL column from highest to lowest
bacteria_data_sorted = bacteria_data.sort_values(by="BSL", ascending=False)
print('Data sorted by descending BSL value:')
display(bacteria_data_sorted)
```

```python
# PART 7: GROUPING ANALYSIS 
# Calculating mean fitness scores grouped by label type to compare bacterial groups.
mean_by_label_type = bacteria_data.groupby("labels")[["cipro_fit", "carb_fit"]].mean().reset_index()
print("Mean of cipro_fit and carb_fit by label type:")
display(mean_by_label_type)
```

```python
# Calculating the maximum fitness scores of cipro_fit and carb_fit grouped by label type.
max_by_label_type = bacteria_data.groupby("labels")[["cipro_fit", "carb_fit"]].max().reset_index()
print("Maximum of cipro_fit and carb_fit by label type:")
display(max_by_label_type)
```
